## Part 2: Answer the following questions

1. **What are the critical parts of the script related on training the Word2Vec?**
**Answer:** The most critical component of the script is the initialization step using Gensim's `Word2Vec` object, specifically inside the `train_sgns()` block. First, supplying the processed/cleaned `sentences` list is essential. Secondly, setting the parameter `sg=1` accurately instructs the model to utilize the **Skip-gram architecture** (as opposed to CBOW which relies on `sg=0`), utilizing context to predict words. Additionally, passing `negative=10` is vital because it explicitly enables the computationally efficient **Negative Sampling** procedure which isolates only a handful of "noise" outputs to update their network weights iteratively instead of the whole massive vocabulary at once on each iteration.

2. **What Word2Vec methods were used to evaluate the similarity of words?**
**Answer:** 
- `model.wv.most_similar(word, topn=N)`: This in-built Gensim framework method directly calculates the distances of all tokens in the vocabulary relative to a specific embedding to find its exact nearest semantic "neighbors" using cosine similarities.
- `model.wv[word]`: This syntax extracts the exact multi-dimensional numpy array encoding representing a single selected string.
- `cosine_similarity(v1, v2)`: Native vector computations using Scikit-Learn that calculate the cosine similarity angles between two isolated tensors (from the aforementioned numpy array extractor) independently to test domain-specific similarity.

```python
import re
import math
import json
import random
from collections import Counter
from typing import List, Tuple, Dict

import requests
from bs4 import BeautifulSoup
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Setup
RANDOM_SEED = 42
random.seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)

def ensure_nltk():
    resources = ["punkt", "punkt_tab", "stopwords"]
    for r in resources:
        try:
            nltk.data.find(f"tokenizers/{r}")
        except LookupError:
            nltk.download(r)

ensure_nltk()
```

```python
# 1. Use a Wikipedia article as dataset
WIKI_URL = "https://en.wikipedia.org/wiki/Valorant"

def fetch_wikipedia_article(url: str) -> str:
    headers = {"User-Agent": "Mozilla/5.0"}
    resp = requests.get(url, headers=headers, timeout=30)
    resp.raise_for_status()
    soup = BeautifulSoup(resp.text, "html.parser")
    content_div = soup.find("div", {"id": "mw-content-text"})
    paragraphs = content_div.find_all(["p", "li"])
    text_blocks = [p.get_text(" ", strip=True) for p in paragraphs if p.get_text(" ", strip=True)]
    text = "\n".join(text_blocks)
    text = re.sub(r"\[[0-9]+\]", " ", text)
    return re.sub(r"\s+", " ", text).strip()

print("Fetching article...")
raw_text = fetch_wikipedia_article(WIKI_URL)
print(f"Fetched {len(raw_text)} characters.")
```

```python
# 2. Preprocess the text
def preprocess_text(text: str) -> List[List[str]]:
    sentences = sent_tokenize(text)
    processed = []
    for sent in sentences:
        sent = sent.lower()
        sent = re.sub(r"[^a-z0-9\-\s]", " ", sent)
        sent = re.sub(r"\s+", " ", sent).strip()
        if not sent: continue
        tokens = word_tokenize(sent)
        cleaned = [tok.strip("-") for tok in tokens if tok.strip("-") and not tok.isdigit() and len(tok) >= 2]
        if len(cleaned) >= 3:
            processed.append(cleaned)
    return processed

sentences = preprocess_text(raw_text)
print(f"Processed {len(sentences)} sentences.")
```

```python
# 3. Train Skip-gram with Negative Sampling
# Parameterized space to test both Window 5 and Window 10
def train_sgns(sentences: List[List[str]], window_size: int):
    model = Word2Vec(
        sentences=sentences,
        vector_size=100, 
        window=window_size,      
        min_count=1,
        workers=4,
        sg=1,         
        negative=10,   # negative sampling
        epochs=200,
        sample=1e-3,
        alpha=0.025,
        min_alpha=0.0007,
        seed=RANDOM_SEED,
    )
    return model

print("Training models...")
model_w5 = train_sgns(sentences, window_size=5)
model_w5.save("model_w5.model")
print(f"Model (Window=5) Vocabulary size: {len(model_w5.wv)} and saved as model_w5.model")

model_w10 = train_sgns(sentences, window_size=10)
model_w10.save("model_w10.model")
print(f"Model (Window=10) Vocabulary size: {len(model_w10.wv)} and saved as model_w10.model")

# Default model variable reserved for PCA block downwards
model = model_w10
```

```python
# 4 & 5. Evaluate and Report Side-by-Side
def has_word(m: Word2Vec, word: str) -> bool:
    return word in m.wv.key_to_index

# Nearest Neighbors Comparison
probe_words = ["riot", "agent", "weapon", "match", "spike"]
print("=== Nearest Neighbors Comparison (Top 3) ===")
for word in probe_words:
    if has_word(model_w5, word) and has_word(model_w10, word):
        print(f"\n{word}:")
        neigh5 = model_w5.wv.most_similar(word, topn=3)
        neigh10 = model_w10.wv.most_similar(word, topn=3)
        print(f"  Window=5:  {[n[0] for n in neigh5]}")
        print(f"  Window=10: {[n[0] for n in neigh10]}")

# Relatedness Test Specific to the Valorant Corpus
relatedness_test = [
    ("riot", "games", 0.95),
    ("tactical", "shooter", 0.95),
    ("agent", "abilities", 0.85),
    ("spike", "plant", 0.90),
    ("weapon", "gun", 0.85),
    ("valorant", "esports", 0.80),
    ("riot", "kitchen", 0.05),
    ("spike", "tractor", 0.02)
]

print("\n=== Relatedness Test Set (Side-by-Side) ===")
print(f"{'Word 1':10s} - {'Word 2':10s} | {'Gold':>4s} | {'W=5 Pred':>8s} | {'W=10 Pred':>9s}")
print("-" * 56)
for w1, w2, gold in relatedness_test:
    if has_word(model_w5, w1) and has_word(model_w10, w2):
        v1_5 = model_w5.wv[w1].reshape(1, -1)
        v2_5 = model_w5.wv[w2].reshape(1, -1)
        pred_5 = float(cosine_similarity(v1_5, v2_5)[0][0])
        
        v1_10 = model_w10.wv[w1].reshape(1, -1)
        v2_10 = model_w10.wv[w2].reshape(1, -1)
        pred_10 = float(cosine_similarity(v1_10, v2_10)[0][0])
        
        print(f"{w1:10s} - {w2:10s} | {gold:.2f} | {pred_5:8.4f} | {pred_10:9.4f}")
```

## 1. Compare the evaluation values to the previous model. What can you derive from the change?

**OLD (window=5):**
The model captured immediate syntactic structures. This means words that frequently appear directly next to each other (e.g., grammatical pairings or strict immediate adjectives like "tactical shooter") yielded higher similarities. The model is constrained to looking at only 5 words backwards and forwards, leading to a narrower scope heavily focused on direct sentence syntax.

**NEW (window=10):**
By doubling the context window, the model drastically captures broader topical and abstract semantic relationships. Words that belong to the same broader theme or topic (e.g., "valorant", "esports", "match") will influence each other significantly more, even if they never sit directly adjacent in a sentence. You generally see higher similarity scores for conceptually related themes across paragraphs, while purely grammatical/adjacent word pair similarity could be diluted slightly since the context window encompasses far more "noise" from the surrounding paragraph text.


```python
# Visualize vectors using PCA (20 known words)
words_to_visualize = [
    "riot", "games", "valorant", "tactical", "shooter", 
    "agent", "abilities", "spike", "plant", "defuse",
    "round", "match", "team", "player", "weapon",
    "gun", "knife", "esports", "map", "site"
]

words_in_vocab = [w for w in words_to_visualize if has_word(model, w)]

if words_in_vocab:
    vectors = np.array([model.wv[w] for w in words_in_vocab])
    pca = PCA(n_components=2)
    vectors_2d = pca.fit_transform(vectors)
    
    plt.figure(figsize=(10, 8))
    plt.scatter(vectors_2d[:, 0], vectors_2d[:, 1], c="blue", edgecolors="k")
    
    for i, word in enumerate(words_in_vocab):
        plt.annotate(word, xy=(vectors_2d[i, 0], vectors_2d[i, 1]), xytext=(5, 2),
                     textcoords="offset points", ha="right", va="bottom", fontsize=12)
        
    plt.title("PCA 2D Visualization of 20 Valorant Domain Word Vectors")
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.show()
else:
    print("None of the specified words were found in the vocabulary.")
```
